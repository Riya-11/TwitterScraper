```python
from bs4 import BeautifulSoup
import re
import nltk
from nltk.corpus import stopwords
import pymongo
from pymongo import MongoClient
import pandas as pd
```

```python
client = MongoClient('mongodb://localhost:27017/')
db = client["tweets"] #Data gets stored in "tweets_spacex" db
collection = db.tweets_spacex
tweets = pd.DataFrame(list(collection.find()))
```

```python
replace_with_space = re.compile('[/(){}\[\]\|@,;]')
deleteSymbol = re.compile('[^0-9a-z #+_]')
stopwords = set(stopwords.words('english'))

def cleanText(text):
    text = BeautifulSoup(text, "lxml").text # HTML decoding
    text = text.lower() # lowercase text
    text = replace_with_space.sub(' ', text) # replace certain symbols by space in text
    text = deleteSymbol.sub('', text) # delete symbols from text
    text = ' '.join(word for word in text.split() if word not in stopwords) # remove stopwords from text
    return text
```

```python
# tweet_text = cleanText(tweets['text'])
tweets['text'] = tweets['text'].apply(cleanText)
```
